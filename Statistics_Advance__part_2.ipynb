###1. What is hypothesis testing in statistics?

####ans :

Hypothesis Testing in Statistics :

Hypothesis testing is a statistical method used to make decisions or draw inferences about a population based on a sample of data. It involves proposing two hypothetical conditions:

1. Null Hypothesis (H₀): A default assumption that there is no significant difference or relationship.
2. Alternative Hypothesis (H₁): A research hypothesis that there is a significant difference or relationship.

Purpose:

Hypothesis testing helps researchers and statisticians:

1. Make informed decisions: Based on data-driven evidence.
2. Test theories: Evaluate the validity of hypotheses.
3. Draw conclusions: About populations based on sample data.



###2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?

####ans :

Null Hypothesis (H0) vs. Alternative Hypothesis (H1) -

Null Hypothesis (H0):

- A statement of no effect or no difference
- Assumes that there is no significant difference or relationship
- Typically denoted as H0

Alternative Hypothesis (H1):

- A statement of an effect or difference
- Assumes that there is a significant difference or relationship
- Typically denoted as H1

Key differences:

1. Statement: H0 states no effect, while H1 states an effect.
2. Direction: H0 assumes no difference, while H1 assumes a difference.
3. Purpose: H0 is tested against H1 to determine if there's enough evidence to reject H0.

Example:

- H0: μ = 0 (The average score is 0)
- H1: μ ≠ 0 (The average score is not 0)

In this example, the null hypothesis assumes no effect (average score is 0), while the alternative hypothesis assumes an effect (average score is not 0).

###3.  What is the significance level in hypothesis testing, and why is it important?

####ans :

Significance Level in Hypothesis Testing -

The significance level, often denoted as α (alpha), is the maximum probability of rejecting the null hypothesis when it is actually true. This is also known as a Type I error.

- Why is it important?

1. Controls Type I error: By setting a significance level, you control the risk of rejecting a true null hypothesis.
2. Determines decision-making: The significance level helps determine whether to reject or fail to reject the null hypothesis.
3. Provides a threshold: α serves as a threshold for determining statistical significance.



###4. What does a P-value represent in hypothesis testing?


####ans :

The P-value represents the probability of obtaining a test statistic as extreme as, or more extreme than, the observed value, assuming that the null hypothesis is true.

Interpretation:

- A small P-value (typically < 0.05) indicates strong evidence against the null hypothesis.
- A large P-value (typically > 0.05) indicates weak evidence against the null hypothesis.

-What does it tell us?

1. Strength of evidence: The P-value indicates the strength of evidence against the null hypothesis.
2. Probability of chance: The P-value represents the probability of observing the test statistic (or a more extreme value) by chance, assuming the null hypothesis is true.



###5. How do you interpret the P-value in hypothesis testing?

####ans :

The P-value is a measure of the strength of evidence against the null hypothesis. Here's how to interpret it:

P-Value Interpretation:

1. P-value ≤ α (significance level): Reject the null hypothesis. There is statistically significant evidence to support the alternative hypothesis.
2. P-value > α (significance level): Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.


###6. What are Type 1 and Type 2 errors in hypothesis testing?

###ans :

Type 1 Error (α-error):

- Rejecting a true null hypothesis: When the null hypothesis is actually true, but the test incorrectly rejects it.
- False positive: A Type 1 error occurs when a test indicates a significant effect or difference when there isn't one.

Type 2 Error (β-error):

- Failing to reject a false null hypothesis: When the null hypothesis is actually false, but the test fails to reject it.
- False negative: A Type 2 error occurs when a test fails to detect a significant effect or difference when there is one.

Consequences:

- Type 1 errors can lead to unnecessary changes or interventions.
- Type 2 errors can lead to missed opportunities or failing to detect important effects.

###9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

###ans :


The Z-score is calculated using the following formula:

z = (x̄ - μ) / (σ / √n)

Where:

- x̄ = sample mean
- μ = population mean
- σ = population standard deviation
- n = sample size

What does the Z-Score represent?

The Z-score represents the number of standard deviations away from the population mean that the sample mean lies. It measures how many standard errors the sample mean is away from the population mean.

Interpretation:

- A Z-score of 0 means the sample mean is equal to the population mean.
- A positive Z-score indicates the sample mean is above the population mean.
- A negative Z-score indicates the sample mean is below the population mean.

In Hypothesis Testing:

The Z-score is used to determine the probability of observing a sample mean as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. This probability is known as the P-value.

- If the P-value is less than the significance level (α), the null hypothesis is rejected, indicating a statistically significant difference between the sample mean and the population mean.

###15. What is the margin of error, and how does it affect the confidence interval?

####ans :

The margin of error is the maximum amount by which the sample estimate may differ from the true population parameter. It's a measure of the uncertainty associated with the estimate.

How it affects the confidence interval:

1. Width of the interval: The margin of error determines the width of the confidence interval.
2. Precision of the estimate: A smaller margin of error indicates a more precise estimate, while a larger margin of error indicates less precision.

Calculating the margin of error:

Margin of error = Critical value × Standard error

Where:

- Critical value depends on the confidence level (e.g., 1.96 for 95% confidence)
- Standard error depends on the sample size and variability

Factors that affect the margin of error:

1. Sample size: Larger samples reduce the margin of error.
2. Variability: Increased variability increases the margin of error.
3. Confidence level: Higher confidence levels increase the margin of error.

Impact on confidence interval:

- Narrower interval: Smaller margin of error results in a narrower confidence interval.
- Wider interval: Larger margin of error results in a wider confidence interval.

Understanding the margin of error helps in interpreting the precision of estimates and constructing confidence intervals.

###16. How is Bayes' Theorem used in statistics, and what is its significance.

1. Update prior beliefs: Incorporate new data to update prior probabilities and obtain posterior probabilities.
2. Make predictions: Use posterior probabilities to make predictions or decisions.

P(A|B) = P(B|A) × P(A) / P(B)

Where:

- P(A|B) = posterior probability of A given B
- P(B|A) = likelihood of B given A
- P(A) = prior probability of A
- P(B) = marginal probability of B

Significance:

1. Incorporates prior knowledge: Bayes' Theorem allows incorporating prior knowledge or experience into statistical models.
2. Updates probabilities: Bayes' Theorem provides a framework for updating probabilities based on new data.
3. Flexible modeling: Bayes' Theorem enables flexible modeling of complex problems.

###1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
#ans :
import numpy as np
from scipy import stats

# Given values
sample_mean = 25.6  # Sample mean
population_mean = 25.0  # Known population mean
population_std = 3.2  # Known population standard deviation
sample_size = 40  # Sample size

# Calculate the standard error
std_error = population_std / np.sqrt(sample_size)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / std_error

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print the results
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

###4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

###5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
#ans :
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_errors(alpha, beta, mu0, mu1, sigma, n):
    # Calculate the critical value
    critical_value = stats.norm.ppf(1 - alpha / 2) * sigma / np.sqrt(n) + mu0

    # Calculate Type 1 error (alpha)
    type1_error = alpha

    # Calculate Type 2 error (beta)
    z_score = (critical_value - mu1) / (sigma / np.sqrt(n))
    type2_error = stats.norm.cdf(z_score)

    # Calculate the power of the test
    power = 1 - type2_error

    return type1_error, type2_error, power

def visualize_errors(alpha, mu0, mu1, sigma, n):
    x0 = np.linspace(mu0 - 3 * sigma / np.sqrt(n), mu0 + 3 * sigma / np.sqrt(n), 100)
    x1 = np.linspace(mu1 - 3 * sigma / np.sqrt(n), mu1 + 3 * sigma / np.sqrt(n), 100)

    y0 = stats.norm.pdf(x0, loc=mu0, scale=sigma / np.sqrt(n))
    y1 = stats.norm.pdf(x1, loc=mu1, scale=sigma / np.sqrt(n))
    critical_value = stats.norm.ppf(1 - alpha / 2) * sigma / np.sqrt(n) + mu0

    plt.plot(x0, y0, label='Null Hypothesis')
    plt.fill_between(x0, y0, where=(x0 < mu0 - stats.norm.ppf(1 - alpha / 2) * sigma / np.sqrt(n)) | (x0 > mu0 + stats.norm.ppf(1 - alpha / 2) * sigma / np.sqrt(n)), alpha=0.5, label='Rejection Region')
    plt.plot(x1, y1, label='Alternative Hypothesis')
    plt.axvline(x=critical_value, color='r', linestyle='--', label='Critical Value')
    plt.axvline(x=mu0 - stats.norm.ppf(1 - alpha / 2) * sigma / np.sqrt(n), color='r', linestyle='--')
    plt.legend()
    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.show()

# Example usage
alpha = 0.05
beta = 0.2
mu0 = 0
mu1 = 1
sigma = 1
n = 100

type1_error, type2_error, power = calculate_errors(alpha, beta, mu0, mu1, sigma, n)
print(f"Type 1 Error: {type1_error:.4f}")
print(f"Type 2 Error: {type2_error:.4f}")
print(f"Power: {power:.4f}")

visualize_errors(alpha, mu0, mu1, sigma, n)
```

###6.  Write a Python program to perform an independent T-test and interpret the results.

```python
#ans :
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data for two groups
np.random.seed(0)
group1 = np.random.normal(loc=20, scale=3, size=30)
group2 = np.random.normal(loc=22, scale=3, size=30)

# Perform the independent T-test
t_stat, p_value = stats.ttest_ind(group1, group2)

# Print the results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The means of the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis. The means of the two groups are not significantly different.")

# Calculate the mean and standard deviation of each group
mean1 = np.mean(group1)
mean2 = np.mean(group2)
std1 = np.std(group1)
std2 = np.std(group2)
print(f"Group 1: mean = {mean1:.2f}, std = {std1:.2f}")
print(f"Group 2: mean = {mean2:.2f}, std = {std2:.2f}")

# Visualize the data
plt.hist(group1, alpha=0.5, label='Group 1')
plt.hist(group2, alpha=0.5, label='Group 2')
plt.legend()
plt.title('Independent T-test')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

```

###7.  Perform a paired sample T-test using Python and visualize the comparison results.

###9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
#ans :
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample, confidence_level=0.95):
    # Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Use sample standard deviation (Bessel's correction)

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(len(sample))

    # Calculate the critical value
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=len(sample) - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(0)
sample = np.random.normal(loc=20, scale=3, size=30)
confidence_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(sample, confidence_level)
print(f"Sample Mean: {np.mean(sample):.4f}")
print(f"{confidence_level*100:.0f}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")


```

###11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process,

```python
#ans :
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the prior distribution
def prior(theta):
    return norm.pdf(theta, loc=0, scale=1)

# Define the likelihood function
def likelihood(data, theta):
    return norm.pdf(data, loc=theta, scale=1)

# Define the posterior distribution using Bayes' Theorem
def posterior(data, theta):
    numerator = likelihood(data, theta) * prior(theta)
    denominator = np.sum(likelihood(data, theta) * prior(theta))
    return numerator / denominator

# Generate some sample data
np.random.seed(0)
true_theta = 2
data = np.random.normal(loc=true_theta, scale=1, size=100)

# Calculate the posterior distribution
theta_values = np.linspace(-5, 5, 1000)
posterior_values = [posterior(data.mean(), theta) for theta in theta_values]

# Normalize the posterior values
posterior_values = np.array(posterior_values) / np.sum(posterior_values)

# Plot the prior and posterior distributions
plt.plot(theta_values, prior(theta_values), label='Prior')
plt.plot(theta_values, posterior_values, label='Posterior')
plt.axvline(x=true_theta, color='r', linestyle='--', label='True theta')
plt.legend()
plt.show()

# Calculate the MAP estimate
map_estimate = theta_values[np.argmax(posterior_values)]
print(f"MAP estimate: {map_estimate:.4f}")
```

###16.  Implement an F-test using Python to compare the variances of two random samples.

```python
#ans :

import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(0)
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=0, scale=2, size=100)

# Calculate the variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Calculate the F-statistic
f_statistic = var1 / var2 if var1 > var2 else var2 / var1

# Calculate the degrees of freedom
df1 = len(sample1) - 1 if var1 > var2 else len(sample2) - 1
df2 = len(sample2) - 1 if var1 > var2 else len(sample1) - 1

# Calculate the p-value
p_value = 2 * min(f.sf(f_statistic, df1, df2), f.cdf(f_statistic, df1, df2))

# Print the results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The variances are not equal.")
else:
    print("Fail to reject the null hypothesis. The variances are equal.")

# Alternatively, use the scipy.stats.f_oneway function for one-sided F-test
from scipy.stats import f_oneway
f_statistic, p_value = f_oneway(sample1, sample2)
print("One-sided F-test:")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

```

###19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
#ans :

import numpy as np
from scipy.stats import shapiro, levene, f_oneway
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    # Normality assumption
    print("Normality Assumption:")
    for i, group in enumerate(groups):
        stat, p_value = shapiro(group)
        print(f"Group {i+1}: p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Warning: Group {i+1} may not be normally distributed.")
        else:
            print(f"Group {i+1} appears to be normally distributed.")

    # Equal variance assumption
    print("\nEqual Variance Assumption:")
    stat, p_value = levene(*groups)
    print(f"p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("Warning: Groups may not have equal variances.")
    else:
        print("Groups appear to have equal variances.")

# Independence assumption
    print("\nIndependence Assumption:")
    print("This assumption cannot be tested statistically. It is assumed based on the study design.")

    # Plot histograms for each group
    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        plt.hist(group, bins=20, alpha=0.5, label=f"Group {i+1}")
        plt.title(f"Group {i+1}")
        plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(0)
group1 = np.random.normal(loc=20, scale=3, size=30)
group2 = np.random.normal(loc=22, scale=3, size=30)
group3 = np.random.normal(loc=25, scale=3, size=30)
check_anova_assumptions(group1, group2, group3)

```

```python
#ans :
import numpy as np
from scipy.stats import chi2

# Set the seed for reproducibility
np.random.seed(0)

# Simulate random data from a normal distribution
mean = 20
variance = 9
sample_size = 100
data = np.random.normal(mean, np.sqrt(variance), sample_size)

# Calculate the sample variance
sample_variance = np.var(data, ddof=1)

# Perform the hypothesis test
null_variance = 10
alpha = 0.05
n = sample_size
df = n - 1
chi2_stat = (n - 1) * sample_variance / null_variance

# Calculate the critical Chi-square value
critical_chi2 = chi2.ppf(1 - alpha / 2, df)

# Calculate the p-value
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

# Print the results
print(f"Sample variance: {sample_variance:.4f}")
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"Critical Chi-square value: {critical_chi2:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis. The population variance is not equal to", null_variance)
else:
    print("Fail to reject the null hypothesis. The population variance is equal to", null_variance)

```

###25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
#ans :
import numpy as np
from scipy.stats import norm

def z_test_proportions(successes1, total1, successes2, total2):
    # Calculate the proportions
    p1 = successes1 / total1
    p2 = successes2 / total2

    # Calculate the pooled proportion
    pooled_p = (successes1 + successes2) / (total1 + total2)

    # Calculate the standard error
    se = np.sqrt(pooled_p * (1 - pooled_p) * (1 / total1 + 1 / total2))

    # Calculate the Z-statistic
    z_stat = (p1 - p2) / se

    # Calculate the p-value
    p_value = 2 * (1 - norm.cdf(np.abs(z_stat)))

    return z_stat, p_value

# Example usage
successes1 = 120  # Number of successes in group 1
total1 = 200  # Total number of observations in group 1
successes2 = 100  # Number of successes in group 2
total2 = 250  # Total number of observations in group 2

z_stat, p_value = z_test_proportions(successes1, total1, successes2, total2)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The proportions are not equal.")
else:
    print("Fail to reject the null hypothesis. The proportions are equal.")

```

###26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
#ans :

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=0, scale=2, size=100)

# Calculate the variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Perform the F-test
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1
df2 = len(data2) - 1
p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

# Print the results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The variances are not equal.")
else:
    print("Fail to reject the null hypothesis. The variances are equal.")

# Visualize the results
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(data1, bins=20, alpha=0.5, label='Data 1')
plt.hist(data2, bins=20, alpha=0.5, label='Data 2')
plt.legend()
plt.title('Histograms of the Data')

plt.subplot(1, 2, 2)
plt.boxplot([data1, data2], labels=['Data 1', 'Data 2'])
plt.title('Boxplots of the Data')
plt.show()

# Plot the F-distribution
x = np.linspace(0, 5, 1000)
y = f.pdf(x, df1, df2)
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.axvline(f_stat, color='red', linestyle='--', label=f'F-statistic = {f_stat:.4f}')
plt.legend()
plt.title('F-Distribution')
plt.show()
```

###27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
#ans :
import numpy as np
from scipy.stats import chisquare, norm
import matplotlib.pyplot as plt

# Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=5, scale=2, size=1000)

# Define the bins for the histogram
bins = np.linspace(0, 10, 11)

# Calculate the observed frequencies
observed_frequencies, _ = np.histogram(data, bins=bins)

# Calculate the expected frequencies under a normal distribution
# Calculate the difference in the cumulative distribution function (CDF) between the bin edges.
expected_frequencies = len(data) * (norm.cdf(bins[1:], loc=5, scale=2) - norm.cdf(bins[:-1], loc=5, scale=2))

# Normalize the expected frequencies so their sum matches the observed frequencies
# This step is crucial to avoid the ValueError in chisquare
expected_frequencies = expected_frequencies * np.sum(observed_frequencies) / np.sum(expected_frequencies)


# Perform the Chi-square test
chi2_stat, p_value = chisquare(observed_frequencies, expected_frequencies)

# Print the results
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The data does not fit the normal distribution.")
else:
    print("Fail to reject the null hypothesis. The data fits the normal distribution.")

# Visualize the results
plt.figure(figsize=(8, 6))
plt.hist(data, bins=bins, alpha=0.5, label='Observed')
plt.plot((bins[:-1] + bins[1:]) / 2, expected_frequencies, 'ro-', label='Expected')
plt.legend()
plt.title('Observed vs Expected Frequencies')
plt.show()
```
