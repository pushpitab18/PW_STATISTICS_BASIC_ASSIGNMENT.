###1. What is a random variable in probability theory?

####ans :

###2. What are the types of random variables?

####ans :

There are two main types of random variables: discrete random variables and continuous random variables. Discrete random variables can take on a countable number of distinct values, such as the number of heads in a series of coin tosses. Continuous random variables can take on any value within a given range, such as the value of a stock price or the height of a person.


###3. What is the difference between discrete and continuous distributions?

####ans :

PDFs are used to calculate probabilities and to understand the properties of a random variable.


###5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

CDFs are used to calculate probabilities and to understand the properties of a random variable. They are particularly useful for calculating probabilities of intervals.

###15. What is the application of Z statistics in hypothesis testing?

###16. How do you calculate a Z-score, and what does it represent?

####ans :

A Z-score is calculated as Z = (X - μ) / σ, where X is the value, μ is the mean, and σ is the standard deviation. The Z-score represents the number of standard deviations a data point is from the mean. A Z-score of 0 means the data point is equal to the mean, while a positive Z-score indicates the data point is above the mean, and a negative Z-score indicates it is below the mean.

###1. Write a Python program to generate a random variable and display its value.


```python
#ans :

import random

def generate_random_variable(min_value, max_value):
    """
    Generate a random variable within a specified range.

    Args:
        min_value (int): The minimum value of the range (inclusive).
        max_value (int): The maximum value of the range (inclusive).

    Returns:
        int: A random variable within the specified range.
    """
    return random.randint(min_value, max_value)

# Example usage
min_value = 1
max_value = 100
random_variable = generate_random_variable(min_value, max_value)
print("Random Variable:", random_variable)

```

###2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
#ans :

import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_pmf(n):
    """
    Calculate the probability mass function (PMF) of a discrete uniform distribution.

    Args:
        n (int): The number of possible outcomes.

    Returns:
        x (numpy array): The possible outcomes.
        probabilities (numpy array): The corresponding probabilities.
    """
    x = np.arange(1, n+1)
    probabilities = np.repeat(1/n, n)
    return x, probabilities

# Example usage
n = 10
x, probabilities = discrete_uniform_pmf(n)

# Plot the PMF
plt.stem(x, probabilities)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.xticks(x)
plt.show()

```

###15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
#ans :

import numpy as np
from scipy.stats import norm

# Sample dataset
sample_mean = 10
sample_std_dev = 2
sample_size = 100

# Population parameters
population_mean = 9

z_score = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f'Z-score: {z_score}')
print(f'p-value: {p_value}')
```

###16.  Create a confidence interval for a dataset using Python and interpret the result.

```python
#ans :

import numpy as np
from scipy.stats import t

# Sample dataset
sample_mean = 10
sample_std_dev = 2
sample_size = 100
confidence_level = 0.95

margin_of_error = t.ppf((1 + confidence_level) / 2, sample_size - 1) * sample_std_dev / np.sqrt(sample_size)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f'Confidence Interval: {confidence_interval}')

```

###17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
#ans :

import numpy as np
from scipy.stats import t

np.random.seed(0)
data = np.random.normal(loc=10, scale=2, size=100)
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)
sample_size = len(data)
confidence_level = 0.95

margin_of_error = t.ppf((1 + confidence_level) / 2, sample_size - 1) * sample_std_dev / np.sqrt(sample_size)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f'Confidence Interval: {confidence_interval}')

```

###23.  Implement Z-score normalization on a dataset and visualize its transformation.
